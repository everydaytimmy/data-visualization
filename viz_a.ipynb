```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python

loan_df = pd.read_csv('./loan_data.csv')
loan_df.head()
```

```python
sns.set()
```

```python
FICO_by_purpose = loan_df.groupby('purpose', as_index=False).fico.mean()
graph = sns.barplot(x='purpose', y='fico', data=FICO_by_purpose)
graph.set(ylim=(650,750))
plt.xticks(rotation=70)
plt.xlabel("Loan Purpose")
plt.ylabel("FICO Score")
FICO_by_purpose
```

```python
rate_by_purpose = loan_df.groupby('purpose', as_index=False)["int.rate"].mean()
graph = sns.barplot(x='purpose', y='int.rate', data=rate_by_purpose)
graph.set(ylim=(.1,.15))

plt.xticks(rotation=70)
plt.xlabel("Loan Purpose")
plt.ylabel("Rate")
rate_by_purpose
```

```python
installment_by_purpose = loan_df.groupby('purpose', as_index=False)["installment"].mean()
graph = sns.barplot(x='purpose', y='installment', data=installment_by_purpose)
graph.set(ylim=(200,500))

plt.xticks(rotation=70)
plt.xlabel("Loan Purpose")
plt.ylabel("Installment")
installment_by_purpose
```

### Conclusions

It was of particular interest to see that despite the FICO scores for being the highest the interest rate was well above the average rate for all of the other loan purposes.
